```python
import numpy as np
from decimal import Decimal, ROUND_HALF_UP

# Carica il dataset
df = pd.read_csv('online_gaming_insights.csv')

# Arrotonda alla seconda cifra decimale
df['PlayTimeHours'] = df['PlayTimeHours'].apply(
    lambda x: float(Decimal(str(x)).quantize(Decimal('0.01'), rounding=ROUND_HALF_UP))
)

df['PlayTimeHours'].head(10)
```

```python
# Seleziona solo le colonne categoriche
categorical_cols = df.select_dtypes(include=['object', 'category'])

# Mostra tutte le categorie e la loro frequenza per ciascuna colonna
for col in categorical_cols:
    print(f"\n--- {col} ---")
    print(df[col].value_counts(dropna=False))

```

```python
# Verifica se ci sono duplicati in PlayerID
duplicati = df[df.duplicated(subset=["PlayerID"], keep=False)]

if duplicati.empty:
    print("Nessun PlayerID duplicato trovato.")
else:
    print(f"Trovati {duplicati['PlayerID'].nunique()} PlayerID duplicati.")
    print("Ecco un'anteprima dei duplicati:")
    print(duplicati.sort_values("PlayerID").head(10))
```

##Divisione del dataset in 2 collection

```python
# Crea il dataset Player
player_columns = [
    "PlayerID",
    "Age",
    "Gender",
    "Location",
    "PlayerLevel",
    "AchievementsUnlocked",
    "EngagementLevel"
]
player_df = df[player_columns].drop_duplicates(subset=["PlayerID"])
player_df
```

```python
# Crea il dataset Session
session_columns = [
    "PlayerID",
    "PlayTimeHours",
    "InGamePurchases",
    "SessionsPerWeek",
    "AvgSessionDurationMinutes",
    "GameGenre",
    "GameDifficulty"
]
session_df = df[session_columns].copy()

# Converte la colonna InGamePurchases in booleano (0 -> False, 1 -> True)
session_df["InGamePurchases"] = session_df["InGamePurchases"].astype(bool)

# Aggiunge SessionID progressivo
session_df.insert(0, "SessionID", range(1, len(session_df) + 1))
session_df
```

```python
# Salva i due nuovi dataset
player_df.to_csv("Player.csv", index=False)
session_df.to_csv("Session.csv", index=False)

print("Player.csv e Session.csv creati con successo!")
```
